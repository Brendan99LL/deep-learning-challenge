## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

application_df = application_df.drop(columns = ['EIN', 'NAME'])
application_df
```

```python
# Determine the number of unique values in each column.

application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"

application_count = application_df['APPLICATION_TYPE'].value_counts()
application_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

cutoff_value = 100
application_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = application_counts[application_counts < cutoff_value].index.tolist()

# Replace in dataframe

for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful

application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"

classification_value_counts = application_df['CLASSIFICATION'].value_counts()
classification_value_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1

classification_value_counts_1 = classification_value_counts.loc[classification_value_counts > 1]
classification_value_counts_1.head()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

class_cutoff_value = 100
#classification_counts = application_df['CLASSIFICATION'].value_counts()
#classifications_to_replace = application_df[classification_counts < class_cutoff_value].index.tolist()
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() < class_cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

categorical_data_dummies = pd.get_dummies(application_df, dtype = int)
categorical_data_dummies
```

```python
# Split our preprocessed data into our features and target arrays

y = categorical_data_dummies["IS_SUCCESSFUL"]
X = categorical_data_dummies.drop(columns = "IS_SUCCESSFUL")

# Split the preprocessed data into a training and testing dataset

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

# **Optimization Attempt # 1**

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

number_input_features = len(X_train_scaled[0])

nn_model_1 = tf.keras.models.Sequential()

# First hidden layer
nn_model_1.add(
    tf.keras.layers.Dense(units = 10, input_dim = number_input_features, activation = "relu")
)

# Second hidden layer
nn_model_1.add(
    tf.keras.layers.Dense(units = 9, activation = "sigmoid")
)

# Curious to see if another layer could increase the accuracy and decrease loss
nn_model_1.add(
    tf.keras.layers.Dense(units = 8, activation = "relu")
)

# Output layer
nn_model_1.add(
    tf.keras.layers.Dense(units = 1, activation = "sigmoid")
)

# Check the structure of the model
nn_model_1.summary()
```

```python
# Compile the model

nn_model_1.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])

# Train the model

fit_model_1 = nn_model_1.fit(X_train_scaled, y_train, epochs = 100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model_1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
